```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Asegúrate de tener df ya cargado desde la Parte IV
ruta_archivo = r'C:\Users\Andy\Documents\Proyecto_1\Data\retail_sales_dataset.csv'
df = pd.read_csv(ruta_archivo)

# 1. Análisis de correlación
numerical_df = df.select_dtypes(include='number')
correlation_matrix = numerical_df.corr()

# 2. Crear un heatmap para visualizar las relaciones entre las variables
plt.figure(figsize=(10, 6))
sns.heatmap(correlation_matrix, annot=True, fmt=".2f", cmap='coolwarm', square=True, cbar_kws={"shrink": .8})
plt.title('Heatmap de Correlaciones', fontsize=18)
plt.show()

# 3. Crear subplots para comparar diferentes variables clave
fig, axes = plt.subplots(2, 2, figsize=(14, 10))

# Subplot 1: Total Amount vs Quantity
sns.scatterplot(data=df, x='Quantity', y='Total Amount', ax=axes[0, 0], color='blue')
axes[0, 0].set_title('Relación entre Cantidad y Total Amount', fontsize=14)
axes[0, 0].set_xlabel('Cantidad')
axes[0, 0].set_ylabel('Total Amount')
axes[0, 0].grid(True)

# Subplot 2: Total Amount vs Price per Unit
sns.scatterplot(data=df, x='Price per Unit', y='Total Amount', ax=axes[0, 1], color='orange')
axes[0, 1].set_title('Relación entre Precio por Unidad y Total Amount', fontsize=14)
axes[0, 1].set_xlabel('Precio por Unidad')
axes[0, 1].set_ylabel('Total Amount')
axes[0, 1].grid(True)

# Subplot 3: Boxplot de Total Amount por Product Category sin palette
sns.boxplot(data=df, x='Product Category', y='Total Amount', ax=axes[1, 0])
axes[1, 0].set_title('Distribución de Total Amount por Categoría de Producto', fontsize=14)
axes[1, 0].set_xlabel('Categoría de Producto')
axes[1, 0].set_ylabel('Total Amount')
axes[1, 0].grid(True)

# Subplot 4: Histograma de Total Amount
sns.histplot(df['Total Amount'], bins=30, kde=True, ax=axes[1, 1], color='green')
axes[1, 1].set_title('Histograma de Total Amount', fontsize=14)
axes[1, 1].set_xlabel('Total Amount')
axes[1, 1].set_ylabel('Frecuencia')
axes[1, 1].grid(True)

# Ajustar el diseño
plt.tight_layout()
plt.show()
```
