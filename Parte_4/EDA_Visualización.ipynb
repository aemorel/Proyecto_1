```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# 1. Cargar los datos
ruta_archivo = r'C:\Users\Andy\Documents\Proyecto_1\Data\retail_sales_dataset.csv'
datos = pd.read_csv(ruta_archivo)

# 2. Análisis Descriptivo
print("\nEstadísticas descriptivas:")
print(datos.describe())

# 3. Visualización
# 3.1 Histograma de Total Amount
plt.figure(figsize=(10, 6))
sns.histplot(datos['Total Amount'], bins=30, kde=True, color='#9080ff') 
plt.title('Histograma de Total Amount')
plt.xlabel('Total Amount')
plt.ylabel('Frecuencia')
plt.grid()
plt.show()

# 3.2 Boxplot de Total Amount
plt.figure(figsize=(10, 6))
sns.boxplot(x=datos['Total Amount'], color='#f46a9b') 
plt.title('Boxplot de Total Amount')
plt.xlabel('Total Amount')
plt.grid()
plt.show()

# 3.3 Gráfico de líneas de Total Amount a lo largo del tiempo
# Convertir la columna 'Date' a tipo datetime
datos['Date'] = pd.to_datetime(datos['Date'])

# Agrupar por fecha y calcular la suma de Total Amount
ventas_diarias = datos.groupby('Date')['Total Amount'].sum().reset_index()

plt.figure(figsize=(14, 7))
plt.plot(ventas_diarias['Date'], ventas_diarias['Total Amount'], marker='o', linestyle='-', color='#9080ff')
plt.title('Tendencia de Ventas Diarias')
plt.xlabel('Fecha')
plt.ylabel('Total Amount')
plt.xticks(rotation=45)
plt.grid()
plt.show()

# 3.4 Gráfico de dispersión entre Price per Unit y Total Amount
plt.figure(figsize=(10, 6))
sns.scatterplot(data=datos, x='Price per Unit', y='Total Amount', hue='Product Category', alpha=0.7, palette='pastel')
plt.title('Relación entre Price per Unit y Total Amount')
plt.xlabel('Price per Unit')
plt.ylabel('Total Amount')
plt.grid()
plt.legend()
plt.show()

# 3.5 Combinación de histogramas y boxplots
# Crear una figura y un conjunto de ejes
fig, axes = plt.subplots(1, 2, figsize=(12, 6))

# Histograma
sns.histplot(datos['Total Amount'], bins=30, kde=True, color='#9080ff', ax=axes[0], label='Histograma')
axes[0].set_title('Distribución de Total Amount (Histograma)')
axes[0].set_xlabel('Total Amount')
axes[0].set_ylabel('Frecuencia')
axes[0].legend()

# Boxplot
sns.boxplot(x=datos['Total Amount'], ax=axes[1], color='#f46a9b')
axes[1].set_title('Distribución de Total Amount (Boxplot)')
axes[1].set_xlabel('Total Amount')

# Ajustar la disposición
plt.tight_layout()
plt.show()
```
